### 1.2 A soft introduction of object oriented programming

Python is an object-oriented programming language, which means our coding logic is based on **objects**. Object in Python is a similar concept as the real-world object. For example, this notebook is an object, the blackboard is an object, you computer is an object. Similarly, the wage1 dataset is an object, the number "1" is an object, and the "Hello World" string is an object. We can categorize objects into different **classes**, so that objects in the same class should share some common features. The documentation of that class would document the **attributes** (properties such as name, length, size ..) and **methods** (what it can do, such as go(), turn(), move()) of all objects that belong to that class.

Besides the example datasets from the wooldridge module, we would encounter a lot of common data files in our daily workflow. Common files name extensions for these data files are *RAW*, *CSV* or *TXT*. Knowing how to import these data files is a critical skill for a data analyst.

### 2.1 Read .csv file

The **pandas** module provides methods for reading various data files.

For example, we can use `pandas.read_csv()` to read files with the `.csv` extension. you can find its [documentation](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) by googling "pandas read csv" .

The pandas.read_csv() *Method* takes more than one **parameters**. These parameters are the options for this method. Values passed to these parameters are called **arguments**. Multiple arguments are separated by `,`. In a document, values following the `=` after the parameter names, if any, are the default values.

Let's read `wage1.csv` with the default arguments.

```python
import pandas as pd
wage1=pd.read_csv("wage1.csv")

"""
Another way to leave comments! Three quotations give you a comment block!

Note the first parameter "filepath_or_buffer=" is not spelled out. When arguments provided
in the documented order, the parameter names and the = signs can be ignored.
"""
```

### 2.2 Pandas DataFrame

pandas **DataFrame** is the most commenly used class (user-defined data type) to store data. The first column of a DataFrame is called **index**, and columns of a DataFrame are of **Series** data type.

To extract column(s) from a DataFrame, the easiest way is to use the **index bracket** `[]`.

```python
print(wage1.columns)
```

```python
wage1["wage"]
```

```python
# The data type of a column is a Series
type(wage1["wage"])
```

If you want to extract more than one columns, pass a **list** into the index bracket `[]`.

```python
wage1[["wage","educ"]]
```

Before we continue, let's "save" the two columns to a variable so that we can reuse it later.

```python
df = wage1[["wage","educ"]]
print(df)
```

### 2.3 Read .txt file

Let's import another .txt data file. Before importing, you can navigate to the `wine.txt` file in side bar and double click to preview the file.

From the preview, we find that this data file provides headers in the first row, and indexes in the first column. Let's pass `0` to the `index_col` parameter. We do not wrap `0` with quation marks because the [documentation](https://pandas.pydata.org/docs/reference/api/pandas.read_table.html) says that the `index_col` parameters accept an **int**.

```python
# import txt with pandas:
wine = pd.read_table("wine.txt",index_col=0)
wine.head()
```

DataFrame also allows you to select a range of rows and columns using **slice object**

```python
# if you prefer to refer to rows and columns with indeces (numbers), use iloc[]
# inside the [], type the range of row indeces of your choice, then a comma, follwed by the range of column indeces
# for example

wine.iloc[0:10, 0:2]

# returns: rows starting at index 0 and stopping right before index 10 (10 not included)
# columns: 0 and 1 (2 not included)
```

```python
# you can also count the index backward

wine.iloc[:, -2:]

# here we extract the last 2 rows.
# -2 means second from the last, : followed by no value means 'to the last'
# the first ':' means 'from first to last'.
```

```python
wine.iloc[[0,2,4],:]
# you can also skip rows or columns by giving a list of indeces
# in this example, we extract row with indeces 0,2, and 4, and we keep all the columns.
```

```python
n = wine.iloc[3,2]
print(n)
```

```python
# when you slice to a single value, it turns into a primitive DataType
type(n)
```
