In the cell below, we introduce four new concepts:

1. instead of presenting the results immmediately, we can store a result by assigning it to a **variable** - *result1* and *result2*

2. **Strings** (text instead of numbers) must be wrapped with quotation marks (you can use either single quotation or double quotation, but if you have quotation inside quotations, you need to use single quotation inside double quotation.)

3. To concatenate two Strings, use the <kbd>+</kbd> sign

4. To insert results into Strings (for example, a sentence), you need to use the **[format()](https://www.w3schools.com/python/ref_string_format.asp)** method. Variable in the parenthesis will be filled into the placeholder {}.

5. If you have two results, use two placeholders, put two variable names in the format method in order and separate them using a comma.

6. Alternatively, you can build a format string by adding letter f before the quotation mark `f""`. You can also format your result using [Python format syntax](https://docs.python.org/3/library/string.html#grammar-token-format-spec-format_spec). In the example below, `.2f` keeps two decimal points and use the fixed-point format

```python
result1 = np.sqrt(16)
print("The square root of 16 is: {}".format(result1))

result2 = "Hello World"
print("Python says " + result2)
```

```python
print("{}, the square root of 16 is: {} ".format(result2, result1))
print(f"{result2}, the square root of 16 is: {result1:.2f} ")
```
